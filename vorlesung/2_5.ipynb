### 2.5 Schätzfunktionen
ZIEL: aus gegebener Stichprobe die Parameter der Verteilung einer Zufallsvariable berechnen

**Maximum-Likelihood-Schätzer**

Voraussetzung: 

Annahme einer Verteilungsfunktion mit dem Parametervektor $ \Theta $; Stichprobe $\left(x_{1} , x_{2}, \ldots , x_{n})\right)$ liegt vor

Likelihood: $L\left(x_{1},\ldots,x_{n};\Theta\right)=\prod^{n}_{i=1}P\left(X=x_{i};\Theta\right)$

logarithmische Likelihood: $ lnL=\sum^{n}_{i=1}P\left(X=x_{i};\Theta\right)$

beste Parameter für Verteilung werden bestimmt mit:
$$
\frac{d\, ln L\left(\Theta\right)}{d\,\Theta}=0 \quad \Rightarrow \quad \Theta = \hat{\Theta}
$$

**Beispiel Normalverteilung**

Verteilungsfunktion
$$
f\left(x_{i}|\mu,\sigma\right)=\frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}}
$$
Likelihood-Funktion
$$
L\left(x_{i}|\mu,\sigma\right)=\left(\frac{1}{\sigma \sqrt{2\pi}}\right)^{n} e^{-\frac{\sum^{n}_{i=1}\left(x_{i}-\mu\right)^{2}}{2\sigma^{2}}}
$$
logarithmierte Likelihood-Funktion
$$
ln L\left(x_{i}|\mu,\sigma\right)=n\cdot ln\left(1\right)- n\cdot ln \left(\sigma \sqrt{2\pi}\right) - \sum^{n}_{i=1}\frac{\left(x_{i} - \mu\right)^{2}}{2\sigma^{2}}
$$

Ableitung nach Parametern
$$
\frac{d\;ln L\left(x_{i}|\mu,\sigma\right)}{d\mu}=\frac{1}{\sigma^{2}}\sum^{n}_{i=1}\left(x_{i}-\mu\right)
$$
$$
\frac{d\;ln L\left(x_{i}|\mu,\sigma\right)}{d\sigma}=-n\frac{\sqrt{2\pi}}{\sigma\cdot\sqrt{2\pi}}-\frac{\left(-2\right)}{2\sigma^{3}}\sum^{n}_{i=1}\left(x_{i}-\mu\right)^{2}
$$
Ableitungen 0 setzen
$$
0=\frac{1}{\sigma^{2}}\sum^{n}_{i=1}\left(x_{i}-\mu\right)
$$
$$
0=-\frac{n}{\sigma}+\frac{1}{\sigma^{3}}\sum^{n}_{i=1}\left(x_{i}-\mu\right)^{2}
$$

nach $\mu$ und $\sigma$ umstellen
$$
0=\sum^{n}_{i=1}\left(x_{i}\right)-n\mu
$$
$$
\mu=\frac{1}{n}\sum^{n}_{i=1}\left(x_{i}\right)=\bar{x}
$$
$$
0=-n+\frac{1}{\sigma^{2}}\sum^{n}_{i=1}\left(x_{i}-\mu\right)^{2}
$$
$$
\sigma^{2}=\frac{1}{n}\sum^{n}_{i=1}\left(x_{i}-\bar{x}\right)^{2}
$$
